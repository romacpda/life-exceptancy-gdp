## Data Import & Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")
```

```python
# Load your dataset
df = pd.read_csv("all_data.csv")

# Quick look
df.head()
```

```python
df.info()
df.describe()
df.columns
df.isnull().sum()
```

```python
df = df.rename(columns={
    'Life expectancy at birth (years)': 'Life_Expectancy'
})
```

## üîç Exploratory Data Analysis (EDA)

In this section, we visually explore trends and patterns in the data to begin answering the project‚Äôs key questions.


### 1. How has life expectancy changed from 2000 to 2015 in each country?

This line plot shows how average life expectancy at birth has changed across six countries.

- Most countries began around 72‚Äì78 years and saw gradual increases.
- **Zimbabwe** stands out with a sharp rise from ~45 to 60 years, starting in 2005.


```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Life_Expectancy', hue='Country', marker='o')
plt.title('Life Expectancy Over Time (2000‚Äì2015)')
plt.ylabel('Life Expectancy (Years)')
plt.xlabel('Year')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

### 2. How has GDP changed from 2000 to 2015?

This plot highlights how the economic power of each country shifted.

- The **United States** has the highest GDP overall, with a steady increase.
- **China** displays the fastest relative growth: a 10x jump.
- **Zimbabwe, Mexico, and Chile** saw more modest GDP changes.


```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='GDP', hue='Country', marker='o')
plt.title('GDP Over Time (2000‚Äì2015)')
plt.ylabel('GDP (USD)')
plt.xlabel('Year')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

### 3. Is there a visible relationship between GDP and life expectancy?

This scatterplot explores the connection between economic output and health.

- A **positive correlation** is visually clear.
- **Zimbabwe** shows steep health gains from low GDP.
- **China** shows wider GDP growth with moderate life expectancy increases.
- **Mexico and Chile** follow tighter, more linear paths.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='GDP', y='Life_Expectancy', hue='Country')
plt.title('GDP vs Life Expectancy')
plt.xlabel('GDP (USD)')
plt.ylabel('Life Expectancy (Years)')
plt.tight_layout()
plt.show()
```

### üß™ Hypothesis Testing: GDP vs Life Expectancy

**Null Hypothesis (H‚ÇÄ):** There is no correlation between GDP and life expectancy.  
**Alternative Hypothesis (H‚ÇÅ):** There is a positive correlation between GDP and life expectancy.


```python
correlation = df['GDP'].corr(df['Life_Expectancy'])
print(f"Correlation between GDP and Life Expectancy: {correlation:.2f}")
```

```python
import seaborn as sns

sns.lmplot(data=df, x='GDP', y='Life_Expectancy', hue='Country', height=6, aspect=1.5)
plt.title("Linear Relationship: GDP vs Life Expectancy")
plt.xlabel("GDP (USD)")
plt.ylabel("Life Expectancy (Years)")
plt.tight_layout()
plt.show()
```

### üß† Interpretation of Correlation

The correlation coefficient between **GDP** and **Life Expectancy** is **0.34**, indicating a **moderate positive relationship**.

This suggests that, across the six countries studied:
- As GDP increases, life expectancy *tends* to rise as well.
- However, the relationship is **not strongly linear**, meaning other factors (like healthcare access, government policy, education, or social stability) may also have a significant impact on life expectancy.

üîç Countries like **Zimbabwe** likely drive this weaker correlation:
- Life expectancy increased dramatically despite relatively low GDP.


```python
country_corr = df.groupby('Country').apply(
    lambda x: x['GDP'].corr(x['Life_Expectancy'])
).reset_index(name='GDP_LifeExp_Correlation')

print(country_corr)
```

### üìä Correlation Between GDP and Life Expectancy (Per Country)

| Country                 | Correlation |
|------------------------|-------------|
| Chile                  | 0.95        |
| China                  | 0.91        |
| Germany                | 0.93        |
| Mexico                 | 0.93        |
| United States of America | 0.98     |
| Zimbabwe               | 0.97        |

These findings show a **consistently strong positive correlation** within each country. This means that, while the overall correlation across all countries was only moderate (0.34), each country individually demonstrates that **as GDP increases, life expectancy reliably improves**.

üí° **Key Insight:**  
The weaker overall correlation


```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Define X and y
X = df[['GDP']].values  # 2D array for sklearn
y = df['Life_Expectancy'].values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Get R¬≤ score
r_squared = model.score(X, y)
print(f"R¬≤ (Coefficient of Determination): {r_squared:.2f}")
```

### üß™ Linear Regression Model: GDP Predicting Life Expectancy (All Countries Combined)

We trained a linear regression model with **GDP** as the independent variable and **life expectancy** as the dependent variable across all six countries (2000‚Äì2015).

**R¬≤ Score:** `0.12`

This means that **only 12%** of the variance in life expectancy can be explained by GDP when looking at all countries together.

üîç **Interpretation:**
- The weak R¬≤ confirms that GDP alone does not strongly predict life expectancy when countries are compared globally.
- This is consistent with our earlier finding that the overall correlation was moderate (0.34).
- It reinforces the importance of analyzing countries **individually**, where we saw R¬≤ values (via correlation) as high as **0.98**.

‚û°Ô∏è **Conclusion:**  
While GDP tends to increase alongside life expectancy, its predictive power is more meaningful **within countries** than **across them**‚Äîhighlighting the impact of other country-specific factors like healthcare systems, education, conflict, and policy.
